```python
import cv2
from ultralytics import YOLO

model = YOLO("best.pt")

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    results = model(frame)

    annotated_frame = results[0].plot()

    cv2.imshow("K3 Detection - Webcam", annotated_frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break
```
